```python
#IVF Latency vs num of queries 

import matplotlib.pyplot as plt

# Table data
num_queries = [10, 50, 100, 500, 1000]
wall_times = [0.0000, 0.0000, 0.0015, 0.0020, 0.0030]
num_vectors = [10, 50, 100, 500, 1000]
latencies = [0.0090, 0.0534, 0.0644, 0.0470, 0.0915]

# Plot settings
fig, ax = plt.subplots()
colors = ['#4682B4', '#FFA500']

# Plot the line graphs
ax.plot(num_queries, wall_times, marker='o', color=colors[0], linestyle='-', label='Faiss')
ax.plot(num_vectors, latencies, marker='o', color=colors[1], linestyle='-', label='Milvus')

# Add labels and title
ax.set_xlabel('Number of Vectors')
ax.set_ylabel('Time (s)')
ax.set_title('Number of Vectors vs Latency')

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add legend
ax.legend()

# Add grid lines
ax.grid(axis='both', linestyle='--', alpha=0.7)

# Add data labels for Faiss (wall times)
for i, v in enumerate(wall_times):
    ax.text(num_queries[i], v + 0.0003, str(v), ha='center')

# Add data labels for Milvus (latencies)
for i, v in enumerate(latencies):
    ax.text(num_vectors[i], v + 0.003, str(v), ha='center')

# Adjust the layout
plt.tight_layout()

# Save the plot as a PDF file
#plt.savefig('build_time_latency_plot.pdf')

# Show the plot
plt.show()
```

```python
#HNSW Latency vs num of queries 

import time
import matplotlib.pyplot as plt

# Wall Time values for Faiss HNSW
num_queries_faiss = [10, 50, 100, 500, 1000]
wall_times_faiss = [0.0000, 0.0006, 0.0005, 0.0019, 0.0071]

# Latency values for Milvus HNSW
num_vectors_milvus = [10, 50, 100, 500, 1000]
latencies_milvus = [0.0110, 0.0525, 0.0643, 0.0840, 0.1309]

# Plot Wall Time for Faiss HNSW
plt.plot(num_queries_faiss, wall_times_faiss, marker='o', linestyle='-', color='b', label='Faiss HNSW')

# Plot Latency for Milvus HNSW
plt.plot(num_vectors_milvus, latencies_milvus, marker='o', linestyle='-', color='r', label='Milvus HNSW')

# Configure the plot
plt.xlabel('Number of Queries / Number of Vectors')
plt.ylabel('Time / Latency (s)')
plt.title('Number of queries vs  Latency')
plt.legend()

# Annotate each data point for Faiss HNSW
for i, j in zip(num_queries_faiss, wall_times_faiss):
    plt.text(i, j, f'{j:.4f}s', ha='center', va='bottom')

# Annotate each data point for Milvus HNSW
for i, j in zip(num_vectors_milvus, latencies_milvus):
    plt.text(i, j, f'{j:.4f}s', ha='center', va='bottom')

plt.show()
```
